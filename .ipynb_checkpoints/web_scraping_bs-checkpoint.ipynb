```python
# ---- Prueba de web scraping con BeautifulSoap sobre Amazon
# ---- Objetivo: ver precios de notebook Asus ROG en Amazon y comparar con Asus.CL 
# ---- Radulescu Rocio

# Importamos librerias necesarias 
#pip install beautifulsoup4
import requests
from bs4 import BeautifulSoup
import pandas as pd 
```

```python
# Inserto header en el request para que no reconozca robot
# https://www.kite.com/python/answers/how-to-make-a-request-with-a-user-agent-in-python#:~:text=Use%20requests.,request%20from%20the%20source%20url%20.
req_amazon = requests.get('https://www.amazon.es/s?k=asus+rog+zephyrus+g14', headers={"User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) Gecko/20100101 Chrome/83.0.4103.97"})
soup_amazon = BeautifulSoup(req_amazon.content)
```

```python
# Busco titulos de articulos en amazon
titulos_note = soup_amazon.findAll(class_ ='a-size-base-plus a-color-base a-text-normal')
list_title = []
for values in titulos_note:
    list_title.append(values.text)
    
# Busco precios de articulos en amazon
precios_note = soup_amazon.findAll(class_ ='a-price-whole')
list_price = []
for values in precios_note:
    list_price.append(values.text)

# Paso lista a df
serie_title = pd.Series(list_title)
serie_price = pd.Series(list_price) 

df = pd.DataFrame ([serie_title,serie_price]).transpose()
df.columns = ['title','price']
df['origen'] = 'Amazon'

df.iloc[0:5]
```

```python
# Agrego informacion de asus.cl
req_falav = requests.get('https://www.falabella.com/falabella-cl/search?Ntt=asus+rog+zephyrus+g14', headers={"User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) Gecko/20100101 Chrome/83.0.4103.97"})
soup_falav = BeautifulSoup(req_falav.content)
```

```python
# Busco titulos de articulos en falavella
titulos_note_f = soup_falav.find_all("b", id=lambda value: value and value.startswith("testId-pod-displaySubTitle"))
                                            
list_title_f = []
for values in titulos_note_f:
    list_title_f.append(values.text)
    
# Busco precios de articulos en amazon
precios_note_f = soup_falav.findAll(class_ ='jsx-1904860942 cmr-icon-container')
list_price_f = []
for values in precios_note_f:
    list_price_f.append(values.text)

# Paso lista a df
serie_title_f = pd.Series(list_title_f)
serie_price_f = pd.Series(list_price_f) 

df_f = pd.DataFrame ([serie_title_f,serie_price_f]).transpose()
df_f.columns = ['title','price']
df_f['origen'] = 'Falabella'

df_f.iloc[0:3]
```

```python
df_total = pd.concat([df[0:5], df_f[0:2]])
df_total
```
